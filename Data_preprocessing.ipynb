## YEAR Analysis

```python
pivoted_year = pd.pivot_table(full_df, values = 'value', index = ['country_region', 'indicator-code', 'sex'], columns = 'year')
pivoted_year
```

```python
for year, values in pivoted_year.items():
    print(year, "", round(values.isna().mean()*100, 2), '% nan')
```

```python
cropped_df = full_df[~full_df['year'].between(1950, 1969)]
```

```python
cropped_df = cropped_df[~cropped_df['year'].between(2021, 2022)]
cropped_df = cropped_df.reset_index()
cropped_df = cropped_df.drop(columns = 'index')
```
